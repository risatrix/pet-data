```python
import datetime 

import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt

pets = pd.read_csv('~/pets/Seattle_Pet_Licenses.csv', encoding='latin-1')
# tried update, but there's a lot of weird missing info in the previous years :(
# pets = pd.read_csv('~/Downloads/Seattle_Pet_Licenses_Oct_2020.csv', encoding='latin-1')
```

```python
# shorten the names and remove punctuation & case for ease of coding
pets = pets.rename(columns={"License Issue Date": "issue_date", "License Number": "license_number", 
                            "Animal's Name": "animal_name", "Species": "species", "Primary Breed": "primary_breed",
                            "Secondary Breed": "secondary_breed", "ZIP Code": "zip",
                           })

# drop duplicates because there are a few
pets = pets.drop_duplicates(['license_number'])

# change issue date to be a datetime field
pets.issue_date = pd.to_datetime(pets.issue_date)

# set index to be date time
pets = pets.sort_values('issue_date').set_index('issue_date')

#remove 2020 because there's only info through January
pets = pets[pets.index.year != 2020]

# add value because every function seems to expect values
pets['value'] = 1
```

```python
year_resample = pets.resample('Y').value.count()

# drop 2020 because there's only data for Jan
# year_resample = year_resample[year_resample.index.year != 2020 ]

fig = plt.figure()
# ax = fig.add_axes([1, 1, 1, 1])
ax = fig.add_subplot(111)
year_resample.plot(kind="bar")

ax.set_xticklabels(year_resample.index.year, rotation=60, fontsize='medium')
ax.set_title('Yearly pet license totals')
ax.set_xlabel('Year')
ax.set_ylabel('Number of pet licenses issued')

ax.set_xticklabels(year_resample.index.year, rotation=60, fontsize='medium')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.gcf().subplots_adjust(bottom=0.25)
plt.gcf().subplots_adjust(left=0.25)
# plt.autoscale()

fig = plt.gcf()
fig.savefig('yearly_license_totals.png', dpi=100)
```

```python
month_int_pivot = pd.pivot_table(pets, index=pets.index.month, columns=pets.index.year,
                    values='value', aggfunc='sum')
month_int_pivot
```

```python
# get set up to plot names
def give_month_name(num):
    month_number = str(num)
    datetime_object = datetime.datetime.strptime(month_number, "%m")
    month_name = datetime_object.strftime("%b") # use lower case for short version
    return month_name

month_numbers = month_int_pivot.index
month_names = month_numbers.map(give_month_name)
```

```python
month_int_pivot.plot()

plt.xticks(month_numbers, month_names, rotation='45')

plt.legend().remove()

ax = plt.gca()

ax.set_xlabel('Month issued')
ax.set_ylabel('Number of pet licenses issued')
ax.set_title('Pet license totals By month, 2000-2019')

ax.annotate('2000-2017', xy=(6,400), xytext=(6,400))
ax.annotate('2018', xy=(6,1100), xytext=(6, 1400))
ax.annotate('2019', xy=(6,1700), xytext=(6,1900))

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.gcf().subplots_adjust(bottom=0.25)

plt.savefig('monthly_license_totals.png', dpi=100)
```

```python
day_int_pivot = pd.pivot_table(pets, index=pets.index.weekday, columns=pets.index.year,
                    values='value', aggfunc='sum')

def give_day_name(num):
    day_number = str(num+1)
    datetime_object = datetime.datetime.strptime(day_number, "%d")
    day_name = datetime_object.strftime("%A") # use lower case for short version
    return day_name

day_numbers = day_int_pivot.index
day_names = day_numbers.map(give_day_name)

day_int_pivot.plot()

plt.xticks(day_numbers, day_names, rotation='45')
plt.legend().remove()

ax = plt.gca()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_xlabel('Day Issued')
ax.set_ylabel('Number of pet licenses issued')
ax.set_title('Pet license totals by day, 2000-2019')

plt.gcf().subplots_adjust(bottom=0.25)

plt.savefig('daily_license_totals.png', dpi=100)
```
